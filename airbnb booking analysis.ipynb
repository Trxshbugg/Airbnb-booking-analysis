# -To find whether there is traffic difference among different areas and finding the reason behind it

```python
ab.head(10)
```

```python
traffic_areas = ab.groupby(["neighbourhood_group", "room_type"])['minimum_nights'].count().reset_index()
traffic_areas= traffic_areas.sort_values(by = 'minimum_nights',ascending = False).head(10)
traffic_areas
```

```python
areas_traffic = traffic_areas["room_type"]
room_stayed = traffic_areas["minimum_nights"]
fig = plt.figure(figsize = (10,5))

plt.bar(areas_traffic, room_stayed, color = 'blue', width = 0.3)
plt.xlabel('Room type')
plt.ylabel('Min night')
plt.title('traffic areas based on min nights booked')
plt.show()
```

# from this visualization we found that most of the people are likely to stay at entire home and private room which are present in manhatten, brooklyn and queens and also visitors prefer to stay in room with listing price less

# - finding out the roomcount in overall NYC acc to listing of room types

```python
plt.rcParams['figure.figsize']=(8,5)
ax= sns.countplot(y='room_type', hue = "neighbourhood_group", data=ab, palette = 'bright')

total = len(ab['room_type'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width()/total)
    x= p.get_x() + p.get_width()+0.2
    y = p.get_y()+ p.get_height()/2
    ax.annotate(percentage, (x,y))

plt.title("count of each room types in NYC")
plt.xlabel("Rooms")
plt.xticks(rotation = 90)
plt.ylabel("room counts")
plt.show()
```

# manhattan has more listed properties with entire home/apt around 27% of total listed properties followed by brooklyn with around 19.6%.

# pvt rooms are more in brooklyn as in 20.7% of the total listed properties followed by manhattan with 16.3% of them. while 6.9% of pvt rooms are from queens.

# we can infer that brooklyn, queens, bronx has more pvt room types while manhattan which has the highest no of listing in entire NYC has more entire room/ apt room types.
